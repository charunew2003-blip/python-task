```python
# Ask the user to input a sentence (representing feedback).
feedback =input("enter the feedback:")
```

```python
# Count how many words are in the sentence.
count=len(feedback.split())
print(f"total words are :{count}")
```

```python

# Check how many times the word “data” appears in the sentence, regardless of case.
data_count = feedback.lower().count("data")
print(f"Count of word 'data' is: {data_count}")
```

```python
# Extract and print the first 5 and last 5 characters of the feedback separately.
first5=feedback[0:5]
last5=feedback[-5:]
print(f"first5 characteristics are:{first5}\n last5 characteristics are:{last5}")
```

```python
 #Reverse the sentence and print it.
 reverse=feedback[::-1]
 print(reverse)
```
