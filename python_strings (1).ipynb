```python
##Day 37 ‚Äì Introduction to Programming Languages and Python
üßë‚Äçüéì Project Title: Student Profile Creator
Scenario:
You are helping to build a basic student information system for a local school.The school wants to store some basic details about each student and let them update their interests.

Tasks:

##Create variables to store the student‚Äôs name, age, and their favourite subjects they are enrolled in.Display the student‚Äôs information in a user-friendly format.Ask the student to enter their favourite subject and add it to their favourite subjects variable.
Display the updated list of subjects.
```

```python
## create varibales
student_name=input("enter your name:",)
student_age=int(input("enter your age:",))
favourite_subjects=["maths","science","social science"]
```

```python
## step 2: display  student information
print("\n---student information--")
print(f"name:{student_name}")
print(f"age:{student_age}")
print(f"favourite_subjects:{',' .join(favourite_subjects)}")
```

```python
## add new  favourite  subjects
new_subject=input("\nenter your  new favourite subjects to add:")
favourite_subjects.append(new_subject)
```

```python
# Step 4: Display updated list
print("\n--- Updated Student Profile ---")
print(f"Name: {student_name}")
print(f"Age: {student_age}")
print(f"Favourite Subjects: {', '.join(favourite_subjects)}")
```
